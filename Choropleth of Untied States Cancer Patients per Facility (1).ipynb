```python
def getChanceOfCancer(age):
    timestep = 5
    data = [0.2, 1.7, 10.3, 30.9, 64.9, 132.9, 199.9, 238.1, 273, 343.9, 428.9, 477.7, 460.2, 416.5, 325]
    index = 0
    #age -= 15
    #total = 3404.1
    
    if(age >= 85):
        return data[-1] / 100000
    
    if(age < 15):
        return 0
    
    while(age > 15):
        age -= 5
        index += 1
        
    return data[index] / 100000
```

```python
import pandas as pd
df = pd.read_csv('sc-est2019-agesex-civ-1.csv')
df = df[["STATE", "NAME", "SEX", "AGE", "POPEST2019_CIV"]]
df = df.loc[df["SEX"] == 2]
df = df.loc[df["NAME"] != "United States"]
```

```python
df.head()
```

```python
for i, j in df.iterrows():
    print(i, j)
    print()
```

```python
casesPerState = []
for ID in range(1, 57):
    if(ID == 11): continue
    state_df = df[df["STATE"] == ID]
    total = 0
    age = 0
    for age_pop in state_df["POPEST2019_CIV"]:
        total += getChanceOfCancer(age) * age_pop
        age += 1
    print(total)
    if total != 0:
        casesPerState.append(total)

casesPerStatePerHospital = [casesPerState[id] / facility_count[id] for id in range(50)]
print(casesPerStatePerHospital)
```

```python
url = (
    "https://raw.githubusercontent.com/python-visualization/folium/master/examples/data"
)
state_geo = f"{url}/us-states.json"
```

```python
state_data = list(zip(ABV, casesPerState))
casesInEachState = pd.DataFrame(state_data, columns = ['state', 'cphps'])
casesInEachState.head()

# initialize the map and store it in a m object
m0 = folium.Map(location=[40, -95], zoom_start=4)

folium.Choropleth(
    geo_data=state_geo,
    name="choropleth",
    data=state_data,
    columns=["state", "cphps"],
    key_on="feature.id",
    fill_color="BuPu",
    fill_opacity=0.7,
    line_opacity=.1,
    legend_name="Expected Number of Cancer Patients",
).add_to(m0)

folium.LayerControl().add_to(m0)

# show the map
m0


```

```python
state_data = list(zip(ABV, facility_count))
facilitiesInEachState = pd.DataFrame(state_data, columns = ['state', 'cphps'])
facilitiesInEachState.head()
# initialize the map and store it in a m object
m1 = folium.Map(location=[40, -95], zoom_start=4)

folium.Choropleth(
    geo_data=state_geo,
    name="choropleth",
    data=state_data,
    columns=["state", "cphps"],
    key_on="feature.id",
    fill_color="PuRd",
    fill_opacity=0.7,
    line_opacity=.1,
    legend_name="Number of Facilities",
).add_to(m1)

folium.LayerControl().add_to(m1)

# show the map
m1
```

```python
state_data = list(zip(ABV, casesPerStatePerHospital))
casesInEachHospital = pd.DataFrame(state_data, columns = ['state', 'cphps'])
casesInEachHospital.head()
# initialize the map and store it in a m object
m2 = folium.Map(location=[40, -95], zoom_start=4)

folium.Choropleth(
    geo_data=state_geo,
    name="choropleth",
    data=state_data,
    columns=["state", "cphps"],
    key_on="feature.id",
    fill_color="YlGn",
    fill_opacity=0.7,
    line_opacity=.1,
    legend_name="Expected Number of Cancer Patients per Facility",
).add_to(m2)

folium.LayerControl().add_to(m2)

# show the map
m2
```

```python
x = np.array(casesPerState)
y = np.array(facility_count)
figure(figsize = (25,12), dpi = 500)
plt.plot(x, y, 'o')
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x+b)
plt.xlabel("Number of Cases")
plt.ylabel("Number of Facilities")
plt.title("Cases of Breast Cancer vs Facilities")

for i, txt in enumerate(ABV):
    plt.annotate(txt, (casesPerState[i], facility_count[i]))
plt.show()
```

```python
final_df = pd.DataFrame(list(zip(ABV, facility_count, casesPerState, casesPerStatePerHospital, [facility_count[id] - (m * casesPerState[id] + b) for id in range(50)])),
columns = ['State', 'Facility count', 'Expected number of cases', 'Expected number of cases per hospital', 'Difference from mean number of facilities'])
final_df.sort_values(by = ['Difference from mean number of facilities'], ascending = True, inplace = True)
final_df.head(50)
```
